```python
#data normalization

df = df.iloc[:,1:]
df_norm = (df - df.mean()) / df.std()
df_norm.head()
```

```python
#select X labels

X = df_norm.iloc[:, :5]
X.head()
```

```python
#select Y labels

Y = df_norm.iloc[:, -1]
Y.head()
```

```python
#Feature and Label values

X_arr = X.values
Y_arr = Y.values
```

```python
X_arr
```

```python
Y_arr
```

```python
#Train and Test Split

X_train, X_test, y_train, y_test = train_test_split(X_arr, Y_arr, test_size = 0.01, shuffle = True, random_state=1)

print('X_train shape: ', X_train.shape)
print('y_train shape: ', y_train.shape)
print('X_test shape: ', X_test.shape)
print('y_test shape: ', y_test.shape)
```

```python
#Create the model
def get_model():

    model = Sequential([
        Dense(10, input_shape = (5,), activation = 'relu'),
        Dense(20, activation = 'relu'),
        Dense(5, activation = 'relu'),
        Dense(1)
    ])

    model.compile(
        loss='mse',
        optimizer='adadelta'
    )

    return model

model = get_model()
model.summary()
```

```python
#Model Training

model = get_model()

# this prediction is before training the model
preds_on_untrained = model.predict(X_test)


# Train model and store in the object history
history = model.fit(
    X_train, y_train,
    validation_data = (X_test, y_test),
    epochs = 1000
)
```

```python
#Plot Training & Validation Loss

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# make predictions on the trained model
preds_on_trained = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def compare_predictions(preds_untrained, preds_trained, true_labels):
    mse_untrained = mean_squared_error(true_labels, preds_untrained)
    mse_trained = mean_squared_error(true_labels, preds_trained)

    mae_untrained = mean_absolute_error(true_labels, preds_untrained)
    mae_trained = mean_absolute_error(true_labels, preds_trained)

    r2_untrained = r2_score(true_labels, preds_untrained)
    r2_trained = r2_score(true_labels, preds_trained)

    print(f"MSE on untrained model: {mse_untrained}")
    print(f"MSE on trained model: {mse_trained}")
    print(f"MAE on untrained model: {mae_untrained}")
    print(f"MAE on trained model: {mae_trained}")
    print(f"R-squared on untrained model: {r2_untrained}")
    print(f"R-squared on trained model: {r2_trained}")

    # Plotting
    plt.figure(figsize=(8, 6))

    plt.scatter(true_labels, preds_untrained, color='blue', label='Untrained Model', alpha=0.6)
    plt.scatter(true_labels, preds_trained, color='green', label='Trained Model', alpha=0.6)

    plt.plot([min(true_labels), max(true_labels)], [min(true_labels), max(true_labels)], '--', color='red', linewidth=2)

    plt.title('Model Predictions vs True Labels')
    plt.xlabel('True Labels')
    plt.ylabel('Predictions')
    plt.legend()

    plt.show()

compare_predictions(preds_on_untrained, preds_on_trained, y_test)
```

```python
#Convert Label Value

y_mean = df['price'].mean()
y_std = df['price'].std()

def convert_label_actual(pred):
    return int(pred * y_std + y_mean)

print(convert_label_actual(-1.836486))
```

```python
#Price Prediction

price_on_untrained = [convert_label_actual(y) for y in preds_on_untrained]
price_on_trained = [convert_label_actual(y) for y in preds_on_trained]
price_y_test = [convert_label_actual(y) for y in y_test]
```

```python
# plot price predictions
compare_predictions(price_on_untrained, price_on_trained, price_y_test)
```
